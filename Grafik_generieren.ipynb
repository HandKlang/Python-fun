```python
# names.csv
# US birth statistics from 1910 to 2014

xs = []
ys = []

counter = []
Anzahl = 0

name = "Annabel"
gender = "F"
state = "CA"
year_from = "1919"
year_to = "2014"

# 123,Anna,1990,F,CA,450
# ID, name, year, gender, state, number

with open("../data/names.csv", "r") as file:
    for line in file:
        
        line_split = line.strip().split(",")
        for year in line:
            if year == "Year":
                continue
    
        if line_split[1] == name and line_split[3] == gender and line_split[4] == state and line_split[2] >= year_from and line_split[2] <= year_to:
            xs.append(int(line_split[2]))
            ys.append(int(line_split[5]))
            counter.append(int(line_split[5]))

for z in counter:
    Anzahl = Anzahl + z

print(Anzahl)

#print(xs)
#print(ys)
```
